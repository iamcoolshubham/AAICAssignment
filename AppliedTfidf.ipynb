```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import os
from tqdm import tqdm
from collections import Counter
from scipy.sparse import csr_matrix
```

```python
corpus = [
     'this is the first document',
     'this document is the second document',
     'and this is the third one',
     'is this the first document',
]
```

# Finding the list of unique words from the corpus

```python
s=list()
l=list()
for i in corpus:
    l=i.split(" ")
    s.extend(l)
s=sorted(s)
unique=set(s)
unique=sorted(unique)
unique=list(unique)
print(unique)
```

```python
vocab={i:j for j,i in enumerate(unique)}
print(vocab)
```

```python
tf=[[0 for i in range(len(unique))]for j in range(len(corpus))]
idf=[[0 for i in range(len(unique))]for j in range(len(corpus))]
tfidf=[[0 for i in range(len(unique))]for j in range(len(corpus))]
```

```python
for i in range(len(corpus)):
    x=corpus[i]
    words=x.split(" ")
    for j in range(len(unique)):
        x=words.count(unique[j])
        if(x>0):
            tf[i][j]=x/len(words)
        
        
```

# Finding inverse document frequency of words in every string

```python
import math
for i in range(len(corpus)):
    x=corpus[i]
    words=x.split(" ")
    for j in range(len(unique)):
        count=0
        if(unique[j] in x):
            for k in (corpus):
                if(unique[j] in k):
                    count+=1
        idf[i][j]=math.log((len(corpus)+1)/(count+1))+1
        
```

# TF-IDF

```python
for i in range(len(tf)):
    for j in range(len(tf[0])):
        tfidf[i][j]=tf[i][j]*idf[i][j]
        
```

```python
import pickle
with open('cleaned_strings', 'rb') as f:
    corpus = pickle.load(f)
```

# Finding unique words in corpus

```python
s=list()
l=list()
for i in corpus:
    l=i.split(" ")
    s.extend(l)
s=sorted(s)
unique=set(s)
unique=sorted(unique)
unique=list(unique)
print(unique)
```

```python
vocab={i:j for j,i in enumerate(unique)}
print(vocab)
```

```python
idf_word=[0 for i in range(len(unique))]
```

# Finding IDF of every unique word in the corpus

```python
for i in range(len(unique)):
    count=0
    for k in corpus:
        if(unique[i] in k):
            count=count+1
            continue
    idf_word[i]=math.log((len(corpus)+1)/(count+1))+1
```

```python
print(idf_word)
idf1=idf_word.copy()
```

# Words with top 50 idf value

```python
max_idf=list()
maxidf_value=list()
count=0
while(count!=50):
    i=idf1.index(max(idf1))
    print(unique[i],max(idf1))
    max_idf.append(unique[i])
    maxidf_value.append(max(idf1))
    idf1[i]=-1
    count=count+1
    
```

```python
print(max_idf)
```

```python
print(maxidf_value)
```

```python
dic={max_idf[i]:maxidf_value[i]for i in range(len(max_idf))} 
print(dic)
```

```python
tf=[[0 for i in range(len(max_idf))]for j in range(len(corpus))]
idf=[[0 for i in range(len(max_idf))]for j in range(len(corpus))]

```

# Term frequency 

```python
for i in range(len(corpus)):
    x=corpus[i]
    words=x.split(" ")
    for j in range(len(max_idf)):
        x=words.count(max_idf[j])
        if(x>0):
            tf[i][j]=x/len(words)
        
        
```

```python
print(tf)
```

```python
print(len(tf))
print(len(tf[0]))
```

# IDF

```python
for i in range(len(corpus)):
    x=corpus[i]
    words=x.split(" ")
    m=0
    for j in max_idf:
        if(j in words):
            idf[i][m]=dic.get(j)
        m=m+1
        
```

```python
for i in idf:
    print(i)
```

```python
print(len(idf))
print(len(idf[0]))
```

# TF-IDF

```python
tfidf=[[0 for i in range(len(max_idf))]for j in range(len(corpus))]
for i in range(len(idf)):
    for j in range(len(idf[0])):
        tfidf[i][j]=tf[i][j]*idf[i][j]
```

```python
for i in tfidf:
    print(i)
```

# Sparse Matrix representation of TF-IDF with top 50 features

```python
print(sparse.csr_matrix(tfidf))
```
